```python
# First we need to load the dataset 
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 

# lets load the data 
df = pd.read_csv("titanic_passengers.csv")

# and show the brielf results before we get too far 
df.head()
```

# Prepping the data

```python
# Map the status of the person to either Died or Survivied 
df['Survived_cat'] = df['Survived'].map({0: 'Died', 1: 'Survived'})
```

# Place the people into age groups 
- 0-12 we will represent as a Child
- 13-17 we will say they are an Teenager
- 18-35 will be a Young Adult
- 36-55 will be Mature Adult
- 55 + will be a Senior
  


```python
ageGroups = [0,12,17,35,55,120]
labels = ['Child', 'Teenager', 'Young Adult', 'Mature Adult', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=ageGroups, labels=labels, right=True)
```

# Now let's ingest the ticket price (fare)

```python
df['FareBin'] = pd.qcut(df['Fare'], 4, labels=['Low','Mid-Low','Mid-High','High'])
```

# Was the person traveling along or with people?
If sinSp + Parch is greater than 0, then they was not alone. 

```python
df['TravelCompanions'] = np.where(df['SibSp'] + df['Parch'] > 0, 'With family', 'Alone')
```

```python
df[['Survived_cat','AgeGroup','FareBin','TravelCompanions']].head()
```


# We create a function so we don't have to type out everything multiple times. 

```python
def show_crosstab(feature):
    print("\nCounts:")
    print(pd.crosstab(df[feature], df['Survived_cat']))
    print("\nPercentages (row-normalized):")
    print(pd.crosstab(df[feature], df['Survived_cat'], normalize='index')*100)
```

# Showing Gender

```python
show_crosstab('Sex')

sns.countplot(data=df, x='Sex', hue='Survived_cat')
plt.title("Survival by Sex (Side-by-Side)")
plt.show()
```

# Yikes those are sad statistics. More women survived than men did.

# Showing Passenger Class

```python
# Replace 1/2/3 with meaningful labels
class_labels = {1: "First Class", 2: "Second Class", 3: "Third Class"}
df['PclassLabel'] = df['Pclass'].map(class_labels)

# Quick check
df[['Pclass','PclassLabel']].head()


show_crosstab('PclassLabel')

sns.countplot(data=df, x='PclassLabel', hue='Survived_cat',
              order=['First Class','Second Class','Third Class'])
plt.title("Survival by Passenger Class")
plt.show()

```

# Wow Majority of the Third Class perished sadly while More First Class passengers survived than perished. 

# Showing Age Groups 

```python
show_crosstab('AgeGroup')

sns.countplot(data=df, x='AgeGroup', hue='Survived_cat',
              order=['Child','Teenager','Young Adult','Mature Adult','Senior'])
plt.title("Survival by Age Group")
plt.show()
```

# Amongst all age groups (excluding children) more people died than survived. 

# Ticket Price Survivals

```python
show_crosstab('FareBin')

sns.countplot(data=df, x='FareBin', hue='Survived_cat',
              order=['Low','Mid-Low','Mid-High','High'])
plt.title("Survival by Fare Bin")
plt.show()
```

# Ticket price seems to play a larger part in if you usurvived. The high fare prices seemed to had more suvivals.

# Lets look at if families survived more than lone people. 

```python
show_crosstab('TravelCompanions')

sns.countplot(data=df, x='TravelCompanions', hue='Survived_cat')
plt.title("Survival by Travel Companions")
plt.show()
```

# Alone passengers seemed to perish more than if they was traveling with someone 

# Now lets look and see if there was a certain suvival rate based on where it embarked from. 

```python
# Replace S, C, Q with full names
embarked_labels = {'S': "Southampton", 'C': "Cherbourg", 'Q': "Queenstown"}
df['EmbarkedLabel'] = df['Embarked'].map(embarked_labels)

# Quick check
df[['Embarked','EmbarkedLabel']].head()



show_crosstab('EmbarkedLabel')

sns.countplot(data=df, x='EmbarkedLabel', hue='Survived_cat',
              order=['Southampton','Cherbourg','Queenstown'])
plt.title("Survival by Port of Embarkation")
plt.show()
```

# Looks like Where a passenger boarded is tied to their class and wealth.

 - Cherbourg had more 1st class passengers → higher survival.

- Southampton had many 3rd class passengers → lower survival.

# Now lets build profiles of who is most likely to survive and who is most likely to perish. 

```python
# Profile A: Most likely to survive
profile_a = df[
    (df['Sex'] == 'female') &
    (df['Pclass'] == 1) &
    (df['AgeGroup'].isin(['Young Adult','Adult'])) &
    (df['FareBin'] == 'High') &
    (df['TravelCompanions'] == 'With family') &
    (df['Embarked'] == 'C')
]
print("Most likely to survive profile:")
print(profile_a[['Name','Sex','Age','Pclass','Fare','TravelCompanions','Embarked','Survived_cat']].head())

# Profile B: Least likely to survive
profile_b = df[
    (df['Sex'] == 'male') &
    (df['Pclass'] == 3) &
    (df['Embarked'] == 'S')
]
print("\nLeast likely to survive profile:")
print(profile_b[['Name','Sex','Age','Pclass','Fare','TravelCompanions','Embarked','Survived_cat']].head())
```
