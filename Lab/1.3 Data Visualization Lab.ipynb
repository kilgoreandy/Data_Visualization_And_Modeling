# Now that we got the data we need, we need to summarize it so we can figure out what makes a good song a SMASH HIT



```python
# Items that we want to summarize 

# Features to summarize
features = ['bpm', 'danceability_%', 'energy_%', 'speechiness_%', 'acousticness_%']

# Get important stats like mean median and percentiles 
summary = df[features].describe(percentiles=[0.25,0.5,0.75,0.9]).T
summary = summary[['mean','50%','25%','75%','90%','min','max']]
summary.rename(columns={'50%':'median','25%':'p25','75%':'p75','90%':'p90'}, inplace=True)
summary.round(2)
```

# So what would make a good smash hit? 
A good smash hit would have the following features:

1. A BPM of 100-140 
2. Uses the Key C# 
3. Uses a Major Mode
4. Needs to be at least 78% dancable or greater. 
5. Needs to have more than 77% of energy in the song. 
6. Needs to have Speechiness of 4% to 11% 
7. Needs Acousticness of more than 25% 

So using the data we have above....... what would make a good smash hit? Lets find out! 

```python
criteria = {
    "bpm_min": 100,
    "bpm_max": 140,
    "key": "C#",
    "mode": "Major",
    "danceability_min": 78,
    "energy_min": 77,
    "speechiness_min": 3,
    "speechiness_max": 12,
    "acousticness_max": 30
}

# JUST SHOVE THE CRITERIA AND LET DF FILTER IT FOR YOU 
top_keys = df['key'].value_counts().head(3).index.tolist()
candidates = df[
    (df['bpm'].between(100, 140)) &
    (df['key'].isin(top_keys)) &       
    (df['mode'] == "Major") &
    (df['danceability_%'] >= 78) &
    (df['energy_%'] >= 77) &
    (df['speechiness_%'].between(3, 12)) &
    (df['acousticness_%'] <= 30)
]
candidates[['artist(s)_name',
            'track_name',
            'bpm',
            'key',
            'mode',
            'danceability_%',
            'energy_%',
            'speechiness_%',
            'acousticness_%',
            'streams']].head(15)




pop_thr = df['streams'].quantile(0.75)
print("Top 25% popularity threshold:", int(pop_thr))

candidates = candidates.copy()
candidates['popular'] = candidates['streams'] >= pop_thr
candidates[['artist(s)_name','track_name','streams','popular']]




```
